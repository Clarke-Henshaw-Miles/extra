```python
import numpy as np
import matplotlib.pyplot as plt
magV, magVerr, BV, BVerr = np.loadtxt("matched_B_V (2).csv", skiprows=1, delimiter=",", usecols=(24, 25, 35, 36), unpack=True)
#magr, Br, Brerr = np.loadtxt("matched_B_r.csv", skiprows=1, delimiter=",", usecols=(24, 35, 36), unpack=True)
```

```python
Vabs = magV - (5 * np.log10(1561)) + 5
#rabs = magr - (5 * np.log10(1561)) + 5
#print(Vabs)
```

```python
plt.plot(BV, Vabs, 'ko')
plt.gca().invert_yaxis()
plt.title("Absolute V Magnitude versus Colour Index")
plt.xlabel("Colour index (B-V)")
plt.ylabel("Absolute magnitude (M_V)")
```

```python
plt.errorbar(BV, Vabs, xerr=BVerr, yerr=magVerr, fmt='x')
plt.gca().invert_yaxis()
plt.title("Absolute V Magnitude versus Colour Index")
plt.xlabel("Colour index (B-V)")
plt.ylabel("Absolute magnitude (M_V)")
```
