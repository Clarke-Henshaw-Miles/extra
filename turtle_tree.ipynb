```python
import numpy as np
import matplotlib.pyplot as plt
magV, magVerr, BV, BVerr = np.loadtxt("matched_B_V (2).csv", skiprows=1, delimiter=",", usecols=(24, 25, 35, 36), unpack=True)
#magr, Br, Brerr = np.loadtxt("matched_B_r.csv", skiprows=1, delimiter=",", usecols=(24, 35, 36), unpack=True)
```

```python
Vabs = magV - (5 * np.log10(1561)) + 5
#rabs = magr - (5 * np.log10(1561)) + 5
#print(Vabs)
```

```python
plt.plot(BV, Vabs, 'ko')
plt.gca().invert_yaxis()
plt.title("Absolute V Magnitude versus Colour Index")
plt.xlabel("Colour index (B-V)")
plt.ylabel("Absolute magnitude (M_V)")
```

```python
plt.errorbar(BV, Vabs, xerr=BVerr, yerr=magVerr, fmt='x')
plt.gca().invert_yaxis()
plt.title("Absolute V Magnitude versus Colour Index")
plt.xlabel("Colour index (B-V)")
plt.ylabel("Absolute magnitude (M_V)")
```

```python
ubv_iso_logAge, ubv_iso_Bmag, ubv_iso_Vmag = np.loadtxt('try3.dat',usecols=(2,27,28),unpack=True)
```

```python
np.unique(ubv_iso_logAge)
```

```python
age1 = np.where(ubv_iso_logAge == 7)
age2 = np.where(ubv_iso_logAge == 7.47712)
age3 = np.where(ubv_iso_logAge == 7.69897)
age4 = np.where(ubv_iso_logAge == 7.8451)
age5 = np.where(ubv_iso_logAge == 7.95424)
```

```python
Bmags_age1 = ubv_iso_Bmag[age1]
Vmags_age1 = ubv_iso_Vmag[age1]
bv1= Bmags_age1 - Vmags_age1
plt.plot(bv1,Vmags_age1,'ro',label='10^7 years' )

Bmags_age2 = ubv_iso_Bmag[age2]
Vmags_age2 = ubv_iso_Vmag[age2]
bv2= Bmags_age2 - Vmags_age2
plt.plot(bv2,Vmags_age2,'yo',label='10^7.47712 years' )

Bmags_age3 = ubv_iso_Bmag[age3]
Vmags_age3 = ubv_iso_Vmag[age3]
bv3= Bmags_age3 - Vmags_age3
plt.plot(bv3,Vmags_age3,'go',label='10^7.69897 years' )

Bmags_age4 = ubv_iso_Bmag[age4]
Vmags_age4 = ubv_iso_Vmag[age4]
bv4= Bmags_age4 - Vmags_age4
plt.plot(bv4,Vmags_age4,'o', color = 'm', label='10^7.8451 years' )

Bmags_age5 = ubv_iso_Bmag[age5]
Vmags_age5 = ubv_iso_Vmag[age5]
bv5= Bmags_age5 - Vmags_age5
plt.plot(bv5,Vmags_age5,'o', color = 'b', label='10^7.95424 years' )

plt.errorbar(BV, Vabs, xerr=BVerr, yerr=magVerr, fmt='x')
plt.gca().invert_yaxis()
plt.xlim(-6,4)
plt.ylim(20, -10)
plt.legend()
```
