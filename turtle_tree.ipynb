```python
import numpy as np
import matplotlib.pyplot as plt
magV, magVerr, BV, BVerr = np.loadtxt("matched_B_V1.csv", skiprows=1, delimiter=",", usecols=(24, 25, 35, 36), unpack=True)
#magr, Br, Brerr = np.loadtxt("matched_B_r.csv", skiprows=1, delimiter=",", usecols=(24, 35, 36), unpack=True)
```

```python
Vabs = magV - (5 * np.log10(1561)) + 5
#rabs = magr - (5 * np.log10(1561)) + 5
#print(Vabs)
```

```python
plt.plot(BV, Vabs, 'ko')
plt.gca().invert_yaxis()
plt.title("Absolute V Magnitude versus Colour Index")
plt.xlabel("Colour index (B-V)")
plt.ylabel("Absolute magnitude (M_V)")
```

```python
plt.errorbar(BV, Vabs, xerr=BVerr, yerr=magVerr, fmt='x')
plt.gca().invert_yaxis()
plt.title("Absolute V Magnitude versus Colour Index")
plt.xlabel("Colour index (B-V)")
plt.ylabel("Absolute magnitude (M_V)")
```

```python
ubv_iso_metal, ubv_iso_Bmag, ubv_iso_Vmag = np.loadtxt('finaltry.dat',usecols=(1,29,30),unpack=True)
```

```python
np.unique(ubv_iso_metal)
```

```python
age1 = np.where(ubv_iso_metal == -1.5)
age2 = np.where(ubv_iso_metal == -1)
age3 = np.where(ubv_iso_metal == -0.5)
age4 = np.where(ubv_iso_metal == 0)
age5 = np.where(ubv_iso_metal == 0.69525)
```

```python
Bmags_age1 = ubv_iso_Bmag[age1]
Vmags_age1 = ubv_iso_Vmag[age1]
bv1= Bmags_age1 - Vmags_age1
plt.plot(bv1,Vmags_age1,'ro',markersize = 3.5, label='M/H = -1.5' )

Bmags_age2 = ubv_iso_Bmag[age2]
Vmags_age2 = ubv_iso_Vmag[age2]
bv2= Bmags_age2 - Vmags_age2
plt.plot(bv2,Vmags_age2,'yo',markersize = 3.5, label='M/H = -1' )

Bmags_age3 = ubv_iso_Bmag[age3]
Vmags_age3 = ubv_iso_Vmag[age3]
bv3= Bmags_age3 - Vmags_age3
plt.plot(bv3,Vmags_age3,'go',markersize = 3.5, label='M/H = -0.5' )

Bmags_age4 = ubv_iso_Bmag[age4]
Vmags_age4 = ubv_iso_Vmag[age4]
bv4= Bmags_age4 - Vmags_age4
plt.plot(bv4,Vmags_age4,'o', color = 'm', markersize = 3.5, label='M/H = 0' )

Bmags_age5 = ubv_iso_Bmag[age5]
Vmags_age5 = ubv_iso_Vmag[age5]
bv5= Bmags_age5 - Vmags_age5
plt.plot(bv5,Vmags_age5,'o', color = 'b', markersize = 3.5, label='M/H = 0.69525' )

plt.errorbar(BV, Vabs, xerr=BVerr, yerr=magVerr, fmt='x', alpha=0.7, label='data with errors')
plt.gca().invert_yaxis()
plt.xlim(-2,4)
plt.ylim(7, -5)
plt.title("Absolute V Magnitude versus Colour Index")
plt.xlabel("Colour index (B-V)")
plt.ylabel("Absolute magnitude (M_V)")
plt.legend()
```
