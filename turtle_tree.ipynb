```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
magV, BV = np.loadtxt("matched_B_V.csv", skiprows=1, delimiter=",", usecols=(24, 35), unpack=True)
magr, Br = np.loadtxt("matched_B_r.csv", skiprows=1, delimiter=",", usecols=(24, 35), unpack=True)
```

```python
plt.plot(BV, magV, 'ko')
plt.gca().invert_yaxis()
plt.title("Absolute V Magnitude versus Colour Index")
plt.xlabel("Colour index (B-V)")
plt.ylabel("Absolute magnitude (M_V)")
```

```python
plt.plot(Br, magr, 'ko')
plt.gca().invert_yaxis()
plt.title("Absolute r Magnitude versus Colour Index")
plt.xlabel("Colour index (B-r)")
plt.ylabel("Absolute magnitude (M_r)")
```
